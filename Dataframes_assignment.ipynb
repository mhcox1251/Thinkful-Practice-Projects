
Find all successful documentary projects and sort them by the amount pledged. Print the top 10 highest pledges.

```
#create a boolean to filter to successful projects:
successful_df = (kickstarters_df['category'] == 'Documentary') & (kickstarters_df['state'] == 'successful')

#Convert back to dataframe format
success = kickstarters_df.iloc[successful_df.values]

#Sort by descending order of pledged amount, print the top 10: 
success.sort_values(by='pledged', ascending=False).head(10)
```

 Create a new column named average_per_backer and set its value to the total amount pledged / number of backers. What happened to the rows with 0 backers? How can this be dealt with?


```
new_column = kickstarters_df.assign(average_per_backer = (kickstarters_df.usd_pledged / kickstarters_df.backers).fillna(value=0))

#Rows with 0 backers returned NaN.

new_column.sort_values(by='backers', ascending=True)
```

Drop all rows with 0 backers then repeat the previous exercise.

```
#Use .dropna() to remove nulls:
no_nan = kickstarters_df.dropna()

new_column = no_nan.assign(average_per_backer = (no_nan.usd_pledged / no_nan.backers))

new_column.head().sort_values(by='average_per_backer')
```
