```
import requests
import pandas
```

```
#initial call

url = 'https://api.harvardartmuseums.org/person'

query = {'apikey': '0b801cc0-04f7-11ea-96a4-79e92a6246ae',
         'page': '1',
         'q': 'culture: British',
         'fields': 'birthplace,displayname'
        }

response = requests.get(url, query)
print('Querying {}'.format(response.url))

data = response.json()

status = response.status_code
print('status: {}'.format(status))

print(data['info'])

total_pages = data['info']['pages']

print(len(data['records']))
print(data['records'])
```

```
british_folk = []
page = 1

while page < total_pages:
  query = {'apikey': '0b801cc0-04f7-11ea-96a4-79e92a6246ae',
         'page': page,
         'q': 'culture: British',
         'fields': 'birthplace,displayname'}
  response = requests.get(url, query)
  if(response.ok):
    print('Querying {}'.format(page))
    if page == 1:
      british_folk =  response.json()['records']
      page = page + 1
    else:
      british_folk = british_folk + response.json()['records']
      page = page + 1
  else:
    print("fix ur query")
```
